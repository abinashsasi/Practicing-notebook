```python
#scope of variable
glob_variable='i am global variable'
def glo():
    print('inside function',glob_variable)
glo()
```

```python
#local variable
def loc():
    loc_variable='i am local variable'
    print(loc_variable)
loc()
```

```python
glo_var="hai h r u"
def var():
    global loc_var
    loc_var='welcome'
    print(loc_var)
    print(glo_var)
var()
print(glo_var)
print(loc_var)
```

```python
var ="happy learning"
def greeting():
    global var
    var="bye bye"
    print(var)
greeting()
print(var)
```

```python
var ="happy learning"
def greeting():
    #global var
    var="python series"
    print(var)
    print(globals()['var'])

greeting()
print(var)
```

```python
#recursion
def greeting():
    print("Gud Morning")
    greeting()
greeting()
```

```python
i=0
def greeting():
    global i
    i+=1
    print("good morning",i)
    greeting()
greeting()
```

```python
#function with recursion
def fact(n):
    if n==0:
        return 1
    return n+fact(n-1)
fact(5)
```
