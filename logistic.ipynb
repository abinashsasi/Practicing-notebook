#Logistic_Regression

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.datasets import load_breast_cancer
dt = load_breast_cancer()
# Read the DataFrame, first using the feature data
data = pd.DataFrame(dt.data, columns=dt.feature_names)
# Add a target column, and fill it with the target data
data['outcome'] = dt.target
# Show the first five rows
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['mean radius'].mean()
```

```python
x=data.iloc[:,:-1]
y=data.outcome
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=10000)
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
#cross tab
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
pd.crosstab(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
model.predict_proba(x_test)
```

```python
y_predict_th3=np.where(model.predict_proba(x_test)[:,1]>0.3,1,0)
```

```python
print(pd.crosstab(y_test,y_predict_th3))
```

```python
def predict_threshold(model,x_test,threshold):
    return np.where(model.predict_proba(x_test)[:,1]>threshold,1,0)
```

```python
from sklearn.metrics import confusion_matrix
for thr in np.arange(0,1.0,0.1):
    y_predict=predict_threshold(model,x_test,thr)
    print('threshold :',thr)
    #print(pd.crosstab(y_test,y_predict))
    print(confusion_matrix(y_test,y_predict))
    print(accuracy_score(y_test,y_predict))

```
