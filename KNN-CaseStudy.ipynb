```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("C:/Users/abina/OneDrive/Desktop/ML/movies.csv")
rating=pd.read_csv("C:/Users/abina/OneDrive/Desktop/ML/ratings.csv")
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].str.split('|')
```

```python
rating.head()
```

```python
finaldata=rating.merge(movies,on='movieId',how='inner')
```

```python
finaldata.head()
```

```python
finaldataset=finaldata.pivot(index='movieId',columns='userId',values='rating')
finaldataset
```

```python
finaldataset.fillna(0,inplace=True)
finaldataset
```

```python
no_user_vote=rating.groupby(['movieId'])['rating'].count()
no_movie_vote=rating.groupby(['userId'])['rating'].count()
```

```python
no_user_vote
no_movie_vote
```

```python
f,ax=plt.subplots(1,1,figsize=(16,4))
plt.scatter(x=no_user_vote.index,y=no_user_vote,color='mediumseagreen')
plt.axhline(y=10,color='r')
plt.xlabel('MovieId')
plt.ylabel('No. of users voted')
plt.show()
```

```python
f,ax=plt.subplots(1,1,figsize=(16,4))
plt.scatter(x=no_movie_vote.index,y=no_movie_vote,color='mediumseagreen')
plt.axhline(y=30,color='r')
plt.xlabel('MovieId')
plt.ylabel('No. of users voted')
plt.show()
```

```python
finaldataset = finaldataset.loc[no_user_vote[no_user_vote > 10].index,:]
print(finaldataset.shape)
final_dataset = finaldataset.loc[:,no_movie_vote[no_movie_vote > 30].index]
final_dataset.shape


```
