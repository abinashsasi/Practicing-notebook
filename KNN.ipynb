```python
df = pd.read_csv('https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/teleCust1000t.csv')
df.head()
```

```python
#df.drop(df[df['custcat'] == 4].index, inplace=True)
df1=df[df['custcat'] != 4]
```

```python
df.nunique()
```

```python
x = df[['region', 'tenure','age', 'marital', 'income', 'ed', 'employ','retire', 'gender']] .values
y = df['custcat'].values
```

```python
#spliting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
x_train_scaled=StandardScaler().fit_transform(x_train)
```

```python
k=6
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train)
model
```

```python
import pickle 
with open("knn.pkl",'wb') as file:
    pickle.dump(model,file)
```

```python
y_predict=model.predict(x_test)
```

#Evaluate accuracy

```python
#cross tab
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
pd.crosstab(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
#train_model
k = 2
best_accuracy=0
k_value=0
#Train Model and Predict
while k<=20:
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(n_neighbors = k).fit(x_train_scaled,y_train)
    

    y_predict=model.predict(x_test)

    #error calucations
    from sklearn.metrics import accuracy_score
    accuracy=(accuracy_score(y_test,y_predict))
    #print(accuracy)
    if accuracy >  best_accuracy:
        best_accuracy = accuracy
        k_value=k
    


    k+=1


print(best_accuracy,"-",k_value)
```
